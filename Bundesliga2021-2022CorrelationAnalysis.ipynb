 # CONCLUSION
 In the correlation analysis, a correlation of 0.941909 between HTAG (Half-Time Away Goals) and FTAG (Full-Time Away Goals) suggests a strong positive relationship between the number of goals scored by the away team during half-time and the number of goals scored by the away team by the end of the match.

In simpler terms, when the away team performs well and scores more goals during the first half of the game, there is a tendency for them to continue performing well and score more goals by the end of the game. Conversely, if the away team struggles to score goals in the first half, they are more likely to have fewer goals by the end of the match.

This high correlation coefficient indicates a consistent pattern where the performance of the away team during the first half of the game has a strong influence on their overall goal-scoring performance in the match.

The comparison of Home Team Shots on Target (HTSOT) with Home Team Shots (HTS) having a correlation coefficient of 0.880487, and Away Team Shots on Target (ATSOT) with Away Team Shots (ATS) having a correlation coefficient of 0.905647, suggests a strong positive relationship between the number of shots on target and the total number of shots for both the home and away teams.

A correlation of 0.880487 between HTSOT and HTS indicates that when the home team takes more shots on target, they tend to have a higher number of total shots. In other words, there is a strong association between the accuracy of the home team's shots and the overall number of attempts they make.

Similarly, a correlation of 0.905647 between ATSOT and ATS suggests that when the away team has more shots on target, they tend to have a higher number of total shots. This implies that the accuracy of the away team's shots is closely related to their overall shooting activity.

Overall, these correlations indicate that both the home and away teams' efficiency in hitting the target (shots on target) is positively correlated with their overall shooting volume (total shots). When teams are more accurate with their shots, they are also more likely to attempt a greater number of shots throughout the match.

The numerical differences between the two correlations, 0.880487 for Home Team Shots on Target (HST) and Home Team Shots (HS), and 0.905647 for Away Team Shots on Target (AST) and Away Team Shots (AS), indicate slightly stronger associations between shots on target and total shots for the away team compared to the home team.

With a correlation of 0.880487 for HST and HS, there is a relatively strong linear relationship between the two variables, indicating that when the home team takes more shots overall, they tend to have a higher number of shots on target.

On the other hand, the correlation of 0.905647 for AST and AS  implies a slightly stronger linear relationship compared to the home team, suggesting that when the away team takes more shots overall, they tend to have a higher number of shots on target.

While both correlations indicate a positive relationship between shots on target and total shots for both teams, the slightly higher correlation for the away team suggests a somewhat stronger association between the accuracy of their shots and the overall number of attempts they make compared to the home team.
